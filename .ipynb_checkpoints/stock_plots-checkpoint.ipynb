beta and alpha can be calculated with numpy. Y = mx + b where x is the base (SPY) and the Symbol compared (AKAM) is y
 m = beta and b = alpha, therefore AKAM plot = beta * returns of SPY + alpha

```python
beta,alpha = np.polyfit(akam_returns['SPY'],akam_returns['AKAM'],1)
plt.plot(akam_returns['SPY'], beta*akam_returns['SPY'] + alpha, '-', color='r')
```
